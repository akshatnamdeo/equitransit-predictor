```python
# Step 1: Data Loading and Initial Exploration

# Importing required libraries
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
import seaborn as sns
import os
from datetime import datetime

# Setting display options for wider output visibility
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
pd.set_option('display.width', 1000)

print("MTA Transit Reliability and Equity Predictor - Data Preprocessing")
print("="*80)

# Loading station mapping data
print("\nLoading Station Mapping Data...")
stations_df = pd.read_csv("datasets/raw/MTA_Subway_Stations.csv")
print(f"Loaded {len(stations_df)} station records")

# Displaying basic details about station dataset
print("\nStation Dataset Overview:")
print(f"Columns: {stations_df.columns.tolist()}")
print(f"Divisions: {stations_df['Division'].unique()}")
print(f"Sample lines: {sorted(stations_df['Line'].unique())[:5]}...")

# Checking for missing values in station dataset
missing_values = stations_df.isnull().sum()
print("\nMissing values per column:")
print(missing_values[missing_values > 0])

# Displaying sample rows from station dataset
print("\nSample of station data:")
stations_df.head()

# Loading delay data
print("\nLoading Delay Data...")
delay_df = pd.read_csv("datasets/raw/MTA_Subway_Trains_Delayed__Beginning_2020_20250331.csv")
print(f"Loaded {len(delay_df)} delay records")

# Displaying basic details about delay dataset
print("\nDelay Dataset Overview:")
print(f"Columns: {delay_df.columns.tolist()}")
print(f"Divisions: {delay_df['division'].unique()}")
print(f"Sample lines: {sorted(delay_df['line'].unique())[:5]}...")
print(f"Unique reporting categories: {delay_df['reporting_category'].nunique()}")
print(f"Unique subcategories: {delay_df['subcategory'].nunique()}")

# Displaying sample rows from delay dataset
print("\nSample of delay data:")
delay_df.head()

# Loading MDBF data
print("\nLoading MDBF Data...")
mdbf_df = pd.read_csv("datasets/raw/MTA_Subway_Mean_Distance_Between_Failures__Beginning_2015_20250331.csv")
print(f"Loaded {len(mdbf_df)} MDBF records")

# Displaying basic details about MDBF dataset
print("\nMDBF Dataset Overview:")
print(f"Columns: {mdbf_df.columns.tolist()}")
print(f"Divisions: {mdbf_df['Division'].unique()}")
print(f"Car Classes: {sorted(mdbf_df['Car Class'].unique())[:5]}...")

# Displaying sample rows from MDBF dataset
print("\nSample of MDBF data:")
mdbf_df.head()

# Sampling ridership data for initial exploration
print("\nSampling Ridership Data...")
ridership_files = glob.glob("datasets/raw/mta_ridership_data (9 GB total) (1176 chunks)/mta_ridership_20250331_093213_chunk*.csv")
sample_ridership = pd.read_csv(ridership_files[0])
print(f"Sampled {len(sample_ridership)} ridership records from first chunk")

# Displaying basic details about ridership dataset
print("\nRidership Dataset Overview:")
print(f"Columns: {sample_ridership.columns.tolist()}")
print(f"Unique stations: {sample_ridership['station_complex'].nunique()}")
print(f"Boroughs: {sample_ridership['borough'].unique()}")

# Checking for missing values in ridership data
missing_values = sample_ridership.isnull().sum()
print("\nMissing values per column:")
print(missing_values[missing_values > 0])

# Displaying sample rows from ridership dataset
print("\nSample of ridership data:")
sample_ridership.head()

# Analyzing date ranges in datasets
print("\nAnalyzing Dataset Time Ranges:")

# Converting date columns to datetime format
delay_df['month_date'] = pd.to_datetime(delay_df['month'])
mdbf_df['month_date'] = pd.to_datetime(mdbf_df['Month'])
sample_ridership['transit_datetime'] = pd.to_datetime(sample_ridership['transit_timestamp'])

# Displaying minimum and maximum date ranges for each dataset
print(f"Delay data: {delay_df['month_date'].min()} to {delay_df['month_date'].max()}")
print(f"MDBF data: {mdbf_df['month_date'].min()} to {mdbf_df['month_date'].max()}")
print(f"Ridership sample: {sample_ridership['transit_datetime'].min()} to {sample_ridership['transit_datetime'].max()}")

# Plotting time coverage of each dataset
plt.figure(figsize=(12, 6))
plt.plot([delay_df['month_date'].min(), delay_df['month_date'].max()], [1, 1], 'r-', linewidth=4, label='Delay Data')
plt.plot([mdbf_df['month_date'].min(), mdbf_df['month_date'].max()], [2, 2], 'b-', linewidth=4, label='MDBF Data')
plt.plot([sample_ridership['transit_datetime'].min(), sample_ridership['transit_datetime'].max()], 
         [3, 3], 'g-', linewidth=4, label='Ridership Sample')

# Finalizing plot settings
plt.yticks([1, 2, 3], ['Delay', 'MDBF', 'Ridership'])
plt.xlabel('Time')
plt.title('Dataset Time Coverage')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig('figures/dataset_time_coverage.png', dpi=300)
plt.show()
```

```python
# Step 2: Division Mapping and Standardization

# Creating mapping for station divisions
division_map_stations = {
    'IRT': 'A',
    'BMT': 'B',
    'IND': 'B',
    'SIR': 'SI'
}

# Creating mapping for delay divisions
division_map_delays = {
    'A DIVISION': 'A',
    'B DIVISION': 'B'
}

# Mapping station divisions
stations_df['Division_Mapped'] = stations_df['Division'].map(division_map_stations)

# Mapping delay divisions
delay_df['division_mapped'] = delay_df['division'].map(division_map_delays)

# Displaying mapping results for verification
print("\nDivision Mapping Results:")

print("\nStation Division Mapping:")
print(stations_df[['Division', 'Division_Mapped']].drop_duplicates())

print("\nDelay Division Mapping:")
print(delay_df[['division', 'division_mapped']].drop_duplicates())

# Checking for unmapped station divisions
unmapped_stations = stations_df[stations_df['Division_Mapped'].isna()]
if len(unmapped_stations) > 0:
    print(f"\nWARNING: {len(unmapped_stations)} station records have unmapped divisions")
    print(unmapped_stations['Division'].value_counts())
else:
    print("\nAll station divisions mapped successfully")

# Checking for unmapped delay divisions
unmapped_delays = delay_df[delay_df['division_mapped'].isna()]
if len(unmapped_delays) > 0:
    print(f"\nWARNING: {len(unmapped_delays)} delay records have unmapped divisions")
    print(unmapped_delays['division'].value_counts())
else:
    print("All delay divisions mapped successfully")

# Verifying consistency of division values across datasets
print("\nVerifying division format consistency with MDBF data:")
print(f"MDBF Division values: {mdbf_df['Division'].unique()}")
print(f"Mapped Station Division values: {stations_df['Division_Mapped'].unique()}")
print(f"Mapped Delay Division values: {delay_df['division_mapped'].unique()}")
```

```python
# Step 3: Line/Route Lookup Table Creation

print("Step 3: Line/Route Lookup Table Creation")
print("="*80)

# Creating route-to-line mapping dictionary from station data
route_to_line = {}

# Iterating through each station record
for _, row in stations_df.iterrows():
    line_name = row['Line']
    if pd.notna(row['Daytime Routes']):
        # Splitting route designators
        routes = str(row['Daytime Routes']).split()
        # Mapping each route to its corresponding line
        for route in routes:
            route_to_line[route] = line_name

# Displaying mapping summary
print(f"Created mapping for {len(route_to_line)} route designators")
print("\nSample of route-to-line mapping:")
for i, (route, line) in enumerate(list(route_to_line.items())[:10]):
    print(f"  Route '{route}' -> Line '{line}'")

# Checking coverage of delay dataset lines
delay_lines = delay_df['line'].unique()
covered_lines = [line for line in delay_lines if line in route_to_line]
coverage_percent = (len(covered_lines) / len(delay_lines)) * 100

# Displaying coverage results
print(f"\nDelay lines coverage: {len(covered_lines)}/{len(delay_lines)} ({coverage_percent:.1f}%)")

# Displaying missing mappings, if any
if len(covered_lines) < len(delay_lines):
    print("\nMissing mappings for these delay lines:")
    for line in sorted(set(delay_lines) - set(covered_lines)):
        print(f"  '{line}'")

# Creating new column for full line names in delay dataset
delay_df['line_full_name'] = delay_df['line'].map(route_to_line)

# Filling missing mappings with route designators
missing_mappings = delay_df['line_full_name'].isna()
if missing_mappings.any():
    print(f"\nFilling in {missing_mappings.sum()} missing line mappings with their route designator")
    delay_df.loc[missing_mappings, 'line_full_name'] = delay_df.loc[missing_mappings, 'line']

# Displaying final mapping results
print("\nFinal mapping results:")
delay_df[['line', 'line_full_name']].drop_duplicates().sort_values('line').head(10)
```

```python
# Step 4: Station Name Standardization

print("Step 4: Station Name Standardization")
print("="*80)

# Importing regex module for name cleaning
import re

# Defining function to remove text in parentheses from station names
def standardize_station_name(name):
    return re.sub(r'\s*\([^)]*\)', '', name).strip()

# Applying standardization to ridership station names
sample_ridership['station_complex_standardized'] = sample_ridership['station_complex'].apply(standardize_station_name)

# Applying standardization to station mapping data
stations_df['stop_name_standardized'] = stations_df['Stop Name'].apply(standardize_station_name)

# Creating set of standardized station names from station mapping
standard_station_names = set(stations_df['stop_name_standardized'])

# Calculating match rate between ridership and station data
ridership_std_stations = set(sample_ridership['station_complex_standardized'])
matching_stations = ridership_std_stations.intersection(standard_station_names)

match_rate = (len(matching_stations) / len(ridership_std_stations)) * 100
print(f"Standardized station name match rate: {match_rate:.1f}% ({len(matching_stations)}/{len(ridership_std_stations)})")

# Displaying sample of standardized names and match results
print("\nSample of standardized station names:")
sample_size = min(5, len(sample_ridership))
for i, row in sample_ridership.head(sample_size).iterrows():
    original = row['station_complex']
    standardized = row['station_complex_standardized']
    match = standardized in standard_station_names
    status = "✓" if match else "✗"
    print(f"  {status} '{original}' -> '{standardized}'")

# Displaying unmatched standardized names with possible matches
if len(matching_stations) < len(ridership_std_stations):
    print("\nSample of standardized names that don't match station data:")
    mismatches = ridership_std_stations - matching_stations
    for name in list(mismatches)[:5]:
        print(f"  '{name}'")
        close_matches = [std for std in standard_station_names if name in std or std in name]
        if close_matches:
            print(f"    Potential matches: {close_matches[:2]}...")

# Creating station-to-lines mapping from standardized station data
station_to_lines = {}

for _, row in stations_df.iterrows():
    station_name = row['stop_name_standardized']
    if pd.notna(row['Daytime Routes']):
        routes = str(row['Daytime Routes']).split()
        if station_name not in station_to_lines:
            station_to_lines[station_name] = set()
        station_to_lines[station_name].update(routes)

# Converting sets to sorted lists
for station, lines in station_to_lines.items():
    station_to_lines[station] = sorted(list(lines))

# Displaying sample of station-to-lines mapping
print("\nSample of station-to-lines mapping:")
for i, (station, lines) in enumerate(list(station_to_lines.items())[:5]):
    print(f"  Station '{station}' -> Lines {lines}")

# Defining function to fetch lines for a given station
def get_lines_for_station(station_name):
    return station_to_lines.get(station_name, [])

# Applying line mapping to ridership dataset
sample_ridership['station_lines'] = sample_ridership['station_complex_standardized'].apply(get_lines_for_station)

# Calculating coverage of station-to-line mapping
stations_with_lines = len([s for s in sample_ridership['station_lines'] if len(s) > 0])
coverage_percent = (stations_with_lines / len(sample_ridership)) * 100
print(f"\nStations with line mappings: {stations_with_lines}/{len(sample_ridership)} ({coverage_percent:.1f}%)")

# Displaying sample of ridership data with line mappings
print("\nSample of ridership data with line mappings:")
sample_ridership[['station_complex', 'station_complex_standardized', 'station_lines']].head()
```

```python
# Step 5: Temporal Alignment
print("Step 5: Temporal Alignment")
print("="*80)

# Reloading original delay data to retain raw date formatting
print("Reloading the original delay data...")
delay_df = pd.read_csv("datasets/raw/MTA_Subway_Trains_Delayed__Beginning_2020_20250331.csv")
print(f"Reloaded {len(delay_df)} delay records")

# Displaying original month values for verification
print("\nOriginal month values in delay_df:")
print(delay_df['month'].head().tolist())

# Parsing date columns
delay_df['month_date'] = pd.to_datetime(delay_df['month'])
mdbf_df['month_date'] = pd.to_datetime(mdbf_df['Month'])
sample_ridership['transit_datetime'] = pd.to_datetime(sample_ridership['transit_timestamp'])

# Displaying parsed date ranges
print("\nDate ranges after parsing:")
print(f"Delay data: {delay_df['month_date'].min()} to {delay_df['month_date'].max()}")
print(f"MDBF data: {mdbf_df['month_date'].min()} to {mdbf_df['month_date'].max()}")
print(f"Ridership sample: {sample_ridership['transit_datetime'].min()} to {sample_ridership['transit_datetime'].max()}")

# Extracting temporal features from each dataset
print("\nExtracting temporal features...")

# Extracting features from delay data
delay_df['year'] = delay_df['month_date'].dt.year
delay_df['month_num'] = delay_df['month_date'].dt.month
delay_df['month_name'] = delay_df['month_date'].dt.strftime('%B')
delay_df['quarter'] = delay_df['month_date'].dt.quarter

# Extracting features from MDBF data
mdbf_df['year'] = mdbf_df['month_date'].dt.year
mdbf_df['month_num'] = mdbf_df['month_date'].dt.month
mdbf_df['month_name'] = mdbf_df['month_date'].dt.strftime('%B')
mdbf_df['quarter'] = mdbf_df['month_date'].dt.quarter

# Extracting features from ridership data
sample_ridership['year'] = sample_ridership['transit_datetime'].dt.year
sample_ridership['month_num'] = sample_ridership['transit_datetime'].dt.month
sample_ridership['day'] = sample_ridership['transit_datetime'].dt.day
sample_ridership['hour'] = sample_ridership['transit_datetime'].dt.hour
sample_ridership['day_of_week'] = sample_ridership['transit_datetime'].dt.dayofweek
sample_ridership['day_name'] = sample_ridership['transit_datetime'].dt.strftime('%A')
sample_ridership['is_weekend'] = sample_ridership['day_of_week'].isin([5, 6])
sample_ridership['is_rush_hour'] = ((sample_ridership['hour'] >= 7) & (sample_ridership['hour'] <= 9)) | \
                                   ((sample_ridership['hour'] >= 16) & (sample_ridership['hour'] <= 19))

# Reapplying division mapping to reloaded delay data
division_map_delays = {
    'A DIVISION': 'A',
    'B DIVISION': 'B'
}
delay_df['division_mapped'] = delay_df['division'].map(division_map_delays)

# Removing target leakage by dropping 'delays' column after creating the target variable
print("\nRemoving 'delays' column to prevent data leakage...")
if 'delays' in delay_df.columns:
    delay_threshold = np.percentile(delay_df['delays'], 65)
    delay_df['significant_delay'] = delay_df['delays'] >= delay_threshold
    print(f"Created target variable 'significant_delay' using threshold: {delay_threshold}")
    delay_df = delay_df.drop(columns=['delays'])
    print("Successfully removed 'delays' column")

# Reapplying line name mapping to delay data
delay_df['line_full_name'] = delay_df['line'].map(route_to_line)
missing_mappings = delay_df['line_full_name'].isna()
if missing_mappings.any():
    delay_df.loc[missing_mappings, 'line_full_name'] = delay_df.loc[missing_mappings, 'line']

# Determining overlapping date range across datasets
delay_min = delay_df['month_date'].min()
delay_max = delay_df['month_date'].max()
mdbf_min = mdbf_df['month_date'].min()
mdbf_max = mdbf_df['month_date'].max()

# Calculating the intersection of time ranges
min_date = max(delay_min, mdbf_min)
max_date = min(delay_max, mdbf_max)

print(f"\nOverlapping time period: {min_date.strftime('%Y-%m-%d')} to {max_date.strftime('%Y-%m-%d')}")

# Filtering datasets to overlapping period
delay_df_filtered = delay_df[(delay_df['month_date'] >= min_date) & (delay_df['month_date'] <= max_date)]
mdbf_df_filtered = mdbf_df[(mdbf_df['month_date'] >= min_date) & (mdbf_df['month_date'] <= max_date)]

# Displaying summary of filtered records
print(f"\nFiltered delay data: {len(delay_df_filtered)}/{len(delay_df)} records ({len(delay_df_filtered)/len(delay_df)*100:.1f}%)")
print(f"Filtered MDBF data: {len(mdbf_df_filtered)}/{len(mdbf_df)} records ({len(mdbf_df_filtered)/len(mdbf_df)*100:.1f}%)")

# Checking ridership data coverage in overlapping time period
ridership_in_range = ((sample_ridership['transit_datetime'] >= min_date) & 
                      (sample_ridership['transit_datetime'] <= max_date))
ridership_coverage = ridership_in_range.mean() * 100
print(f"Ridership sample in overlapping period: {ridership_in_range.sum()}/{len(sample_ridership)} records ({ridership_coverage:.1f}%)")

# Displaying enhanced delay data sample
print("\nSample of delay data with temporal features:")
print(delay_df_filtered[['month', 'month_date', 'year', 'month_num', 'month_name', 'quarter',
                         'division_mapped', 'line', 'line_full_name', 'significant_delay']].head())
```

```python
# Step 6: Ridership Data Processing with Enhanced Station Support
print("Step 6: Ridership Data Processing with Enhanced Station Support")
print("="*80)

# Creating timestamp-based folder for output
import datetime
timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
output_folder = f"datasets/processed"

import os
os.makedirs(output_folder, exist_ok=True)
print(f"Created output folder: {output_folder}")

# Displaying available columns in the ridership data
print("Available columns in sample_ridership:")
print(sample_ridership.columns.tolist())

# Displaying sample values from transit_timestamp column
print("\nSample transit_timestamp values:")
print(sample_ridership['transit_timestamp'].head())

# Extracting date components from transit_timestamp
print("\nExtracting date components from transit_timestamp...")
sample_ridership['transit_timestamp'] = pd.to_datetime(sample_ridership['transit_timestamp'])
sample_ridership['year'] = sample_ridership['transit_timestamp'].dt.year
sample_ridership['month_num'] = sample_ridership['transit_timestamp'].dt.month
sample_ridership['day'] = sample_ridership['transit_timestamp'].dt.day
sample_ridership['day_of_week'] = sample_ridership['transit_timestamp'].dt.dayofweek
sample_ridership['is_weekend'] = sample_ridership['day_of_week'].isin([5, 6])

print("\nAggregating ridership data...")

# Aggregating data by station and date
station_daily_ridership = sample_ridership.groupby(
    ['station_complex_standardized', 'year', 'month_num', 'day']
).agg({
    'ridership': ['sum', 'mean', 'std', 'count'],
    'transfers': ['sum', 'mean'],
    'is_weekend': 'first',
    'borough': 'first'
}).reset_index()

# Flattening column names
station_daily_ridership.columns = ['_'.join(col).strip('_') if isinstance(col, tuple) else col for col in station_daily_ridership.columns]

# Calculating additional metrics
station_daily_ridership['ridership_per_hour'] = station_daily_ridership['ridership_sum'] / station_daily_ridership['ridership_count']
station_daily_ridership['transfer_ratio'] = station_daily_ridership['transfers_sum'] / station_daily_ridership['ridership_sum'].replace(0, 1)

# Creating full date column for easier merges
station_daily_ridership['full_date'] = pd.to_datetime(
    station_daily_ridership['year'].astype(str) + '-' +
    station_daily_ridership['month_num'].astype(str).str.zfill(2) + '-' +
    station_daily_ridership['day'].astype(str).str.zfill(2)
)

# Aggregating data at station level
station_metrics = sample_ridership.groupby('station_complex_standardized').agg({
    'ridership': ['mean', 'std', 'sum', 'max'],
    'transfers': ['mean', 'sum'],
    'borough': 'first',
    'latitude': 'first',
    'longitude': 'first'
}).reset_index()

# Flattening column names
station_metrics.columns = ['_'.join(col).strip('_') if isinstance(col, tuple) else col for col in station_metrics.columns]

# Calculating variation and ratios
station_metrics['ridership_cv'] = station_metrics['ridership_std'] / station_metrics['ridership_mean'].replace(0, 1)
station_metrics['transfer_rate'] = station_metrics['transfers_sum'] / station_metrics['ridership_sum'].replace(0, 1)

# Creating mapping from station to lines using parentheses in raw station name
station_to_lines_mapping = {}
for idx, row in sample_ridership.iterrows():
    station = row['station_complex_standardized']
    line_match = re.search(r'\(([A-Za-z0-9,]+)\)', row['station_complex'])
    if line_match:
        lines = line_match.group(1).split(',')
        if station not in station_to_lines_mapping:
            station_to_lines_mapping[station] = set()
        station_to_lines_mapping[station].update(lines)

# Converting sets to lists
for station in station_to_lines_mapping:
    station_to_lines_mapping[station] = list(station_to_lines_mapping[station])

# Adding line info to station metrics
station_metrics['station_lines'] = station_metrics['station_complex_standardized'].map(station_to_lines_mapping)
station_metrics['station_lines'] = station_metrics['station_lines'].apply(lambda x: x if isinstance(x, list) else [])
station_metrics['num_lines'] = station_metrics['station_lines'].apply(len)
station_metrics['is_transfer_station'] = station_metrics['num_lines'] > 1
station_metrics['primary_line'] = station_metrics['station_lines'].apply(lambda x: x[0] if len(x) > 0 else None)

# Creating geometry points for spatial operations
from shapely.geometry import Point
station_metrics['geometry'] = station_metrics.apply(
    lambda row: Point(row['longitude_first'], row['latitude_first']),
    axis=1
)

# Converting to GeoDataFrame for spatial use
import geopandas as gpd
station_gdf = gpd.GeoDataFrame(
    station_metrics,
    geometry='geometry',
    crs="EPSG:4326"
)

# Displaying final summaries
print(f"\nCreated daily ridership aggregations for {len(station_daily_ridership)} station-days")
print(f"Created station-level metrics for {len(station_metrics)} stations")
print(f"Created GeoDataFrame with {len(station_gdf)} station locations")

# Displaying samples
print("\nSample of daily ridership aggregations:")
print(station_daily_ridership.head())

print("\nSample of station-level metrics:")
print(station_metrics.head())

# Displaying sample of station-to-lines mapping
print("\nSample of station-to-lines mapping:")
sample_mapping = {k: v for i, (k, v) in enumerate(station_to_lines_mapping.items()) if i < 5}
for station, lines in sample_mapping.items():
    print(f"  {station}: {lines}")

# Saving outputs
# station_daily_ridership.to_csv(f"{output_folder}/station_daily_ridership.csv", index=False)
# station_metrics.to_csv(f"{output_folder}/station_metrics.csv", index=False)
# station_gdf.to_file(f"{output_folder}/station_geometry.gpkg", driver="GPKG")
```

```python
# Step 7: Demographic Data Integration
print("Step 7: Demographic Data Integration")
print("="*80)

# Loading NYC census tract shapefile
print("Loading NYC census tract shapefile...")
import geopandas as gpd
nyct_gdf = gpd.read_file("datasets/raw/nyc demographics/nyct2020_25a/nyct2020.shp")
print(f"Loaded {len(nyct_gdf)} census tracts")
print("Census tract data sample:")
print(nyct_gdf.head())

# Loading demographic datasets
print("\nLoading demographic datasets...")

# Defining function to load each demographic dataset
def load_demographic_dataset(base_name):
    metadata_path = f"datasets/raw/nyc demographics/ACSDT5Y2023.{base_name}-Column-Metadata.csv"
    metadata_df = pd.read_csv(metadata_path)
    col_mapping = dict(zip(metadata_df['Column Name'], metadata_df['Label']))
    
    data_path = f"datasets/raw/nyc demographics/ACSDT5Y2023.{base_name}-Data.csv"
    data_df = pd.read_csv(data_path, skiprows=1)
    
    return data_df, col_mapping

# Specifying datasets to load
demographic_datasets = {
    'income': 'B19013',
    'race': 'B02001',
    'hispanic_origin': 'B03002',
    'vehicles': 'B08201',
    'commute_mode': 'B08301',
    'employment': 'B23025'
}

# Loading datasets and storing mappings
demo_data = {}
demo_mappings = {}
for key, base_name in demographic_datasets.items():
    print(f"Loading {key} data...")
    data_df, col_mapping = load_demographic_dataset(base_name)
    demo_data[key] = data_df
    demo_mappings[key] = col_mapping
    print(f"  {len(data_df)} tracts loaded with {len(data_df.columns)} variables")

# Extracting and computing key demographic indicators
print("\nExtracting key demographic indicators...")
census_demographics = pd.DataFrame({'Geography': demo_data['income']['Geography']})
census_demographics['Geographic_Area_Name'] = demo_data['income']['Geographic Area Name']
census_demographics['GEOID'] = census_demographics['Geography'].str.replace('1400000US', '')

# Adding income
income_col = 'Estimate!!Median household income in the past 12 months (in 2023 inflation-adjusted dollars)'
census_demographics['median_income'] = pd.to_numeric(demo_data['income'][income_col], errors='coerce')

# Adding race percentages
race_data = demo_data['race']
census_demographics['total_population'] = pd.to_numeric(race_data['Estimate!!Total:'], errors='coerce')
census_demographics['pct_white'] = pd.to_numeric(race_data['Estimate!!Total:!!White alone'], errors='coerce') / census_demographics['total_population'] * 100
census_demographics['pct_black'] = pd.to_numeric(race_data['Estimate!!Total:!!Black or African American alone'], errors='coerce') / census_demographics['total_population'] * 100
census_demographics['pct_asian'] = pd.to_numeric(race_data['Estimate!!Total:!!Asian alone'], errors='coerce') / census_demographics['total_population'] * 100

# Adding Hispanic/Latino origin
hisp_data = demo_data['hispanic_origin']
census_demographics['pct_hispanic'] = pd.to_numeric(hisp_data['Estimate!!Total:!!Hispanic or Latino:'], errors='coerce') / pd.to_numeric(hisp_data['Estimate!!Total:'], errors='coerce') * 100

# Adding vehicle availability
vehicle_data = demo_data['vehicles']
census_demographics['total_households'] = pd.to_numeric(vehicle_data['Estimate!!Total:'], errors='coerce')
census_demographics['pct_no_vehicle'] = pd.to_numeric(vehicle_data['Estimate!!Total:!!No vehicle available'], errors='coerce') / census_demographics['total_households'] * 100
census_demographics['pct_one_vehicle'] = pd.to_numeric(vehicle_data['Estimate!!Total:!!1 vehicle available'], errors='coerce') / census_demographics['total_households'] * 100
census_demographics['pct_two_plus_vehicles'] = pd.to_numeric(vehicle_data['Estimate!!Total:!!2 vehicles available'], errors='coerce') / census_demographics['total_households'] * 100

# Adding commute data
commute_data = demo_data['commute_mode']
census_demographics['total_commuters'] = pd.to_numeric(commute_data['Estimate!!Total:'], errors='coerce')
census_demographics['pct_public_transit'] = pd.to_numeric(commute_data['Estimate!!Total:!!Public transportation (excluding taxicab):'], errors='coerce') / census_demographics['total_commuters'] * 100
census_demographics['pct_drive_alone'] = pd.to_numeric(commute_data['Estimate!!Total:!!Car, truck, or van:!!Drove alone'], errors='coerce') / census_demographics['total_commuters'] * 100
census_demographics['pct_carpool'] = pd.to_numeric(commute_data['Estimate!!Total:!!Car, truck, or van:!!Carpooled:'], errors='coerce') / census_demographics['total_commuters'] * 100
census_demographics['pct_walk'] = pd.to_numeric(commute_data['Estimate!!Total:!!Walked'], errors='coerce') / census_demographics['total_commuters'] * 100
census_demographics['pct_work_from_home'] = pd.to_numeric(commute_data['Estimate!!Total:!!Worked from home'], errors='coerce') / census_demographics['total_commuters'] * 100

# Adding employment data
employment_data = demo_data['employment']
census_demographics['labor_force_participation'] = pd.to_numeric(employment_data['Estimate!!Total:!!In labor force:'], errors='coerce') / pd.to_numeric(employment_data['Estimate!!Total:'], errors='coerce') * 100
census_demographics['unemployment_rate'] = pd.to_numeric(employment_data['Estimate!!Total:!!In labor force:!!Civilian labor force:!!Unemployed'], errors='coerce') / pd.to_numeric(employment_data['Estimate!!Total:!!In labor force:'], errors='coerce') * 100

# Cleaning up NaNs
census_demographics = census_demographics.fillna(0)

# Merging census demographics with shapefile
print("\nMerging demographic data with census tract geometries...")
nyct_with_demographics = nyct_gdf.merge(census_demographics, on='GEOID', how='left')

# Checking data coverage
data_coverage = (nyct_with_demographics['median_income'].notna().sum() / len(nyct_with_demographics)) * 100
print(f"Census tracts with complete demographic data: {data_coverage:.1f}%")

# Creating 500m catchment areas around stations
print("\nCalculating station catchment areas...")
station_gdf_projected = station_gdf.to_crs("EPSG:2263")
nyct_with_demographics_projected = nyct_with_demographics.to_crs("EPSG:2263")
station_buffers = station_gdf_projected.copy()
station_buffers['buffer_geometry'] = station_buffers.geometry.buffer(500)
station_buffers.set_geometry('buffer_geometry', inplace=True)

# Performing spatial join
print("Performing spatial joins between stations and census tracts...")
station_tract_intersections = gpd.sjoin(
    station_buffers,
    nyct_with_demographics_projected,
    how="left",
    predicate="intersects"
)

# Aggregating demographic data by station
print("Aggregating demographic data by station...")
demographic_cols = [
    'median_income', 'total_population', 'pct_white', 'pct_black', 'pct_asian',
    'pct_hispanic', 'pct_no_vehicle', 'pct_public_transit', 'unemployment_rate',
    'labor_force_participation', 'pct_walk', 'pct_work_from_home', 'total_households'
]

# Converting any non-numeric columns
for col in demographic_cols:
    if col in station_tract_intersections.columns and not pd.api.types.is_numeric_dtype(station_tract_intersections[col]):
        station_tract_intersections[col] = pd.to_numeric(station_tract_intersections[col], errors='coerce')

# Computing mean values by station
available_demo_cols = [col for col in demographic_cols if col in station_tract_intersections.columns]
station_demographics = station_tract_intersections.groupby('station_complex_standardized')[available_demo_cols].mean().reset_index()
print(f"Created demographics for {len(station_demographics)} stations")

# Merging with station metrics
station_metrics_with_demographics = station_metrics.merge(station_demographics, on='station_complex_standardized', how='left')

# Handling missing values
print("\nHandling missing values in final dataset...")
for col in available_demo_cols:
    if station_metrics_with_demographics[col].isna().sum() > 0:
        if pd.api.types.is_numeric_dtype(station_metrics_with_demographics[col]):
            median_val = station_metrics_with_demographics[col].median()
            station_metrics_with_demographics[col].fillna(median_val, inplace=True)

# Saving outputs
print("\nSaving demographic results...")
station_metrics_with_demographics.to_csv(f"{output_folder}/station_demographics.csv", index=False)
nyct_with_demographics.to_file(f"{output_folder}/census_tracts_with_demographics.gpkg", driver="GPKG")

print(f"Demographic data processing complete and saved to {output_folder}")
```

```python
# Step 8: Creating the Daily Target Variable with Station Alignment

from tqdm import tqdm

print("Step 8: Creating the Daily Target Variable with Station Alignment")
print("=" * 80)

# Converting month strings to datetime objects
print("Loading and analyzing daily delay data...")
delay_df['month_date'] = pd.to_datetime(delay_df['month'])

# Defining function for distributing monthly delays to individual days
def distribute_to_daily(row):
    """Distributing monthly significant delays across days"""
    month_start = row['month_date']
    next_month = pd.Timestamp(month_start.year + (month_start.month == 12), (month_start.month % 12) + 1, 1)
    days_in_month = (next_month - month_start).days

    # Generating dates for all days in the month
    dates = [month_start + pd.Timedelta(days=i) for i in range(days_in_month)]

    # Creating random weights for days
    np.random.seed(sum(map(ord, str(row['line']) + str(row['month_date']))))
    weights = np.random.dirichlet(np.ones(days_in_month))

    # Determining number of significant delay days
    if row['significant_delay']:
        num_delay_days = max(1, int(days_in_month * 0.65))
    else:
        num_delay_days = max(1, int(days_in_month * 0.25))

    # Selecting top days based on weights
    day_indices = np.argsort(weights)[-num_delay_days:]

    # Creating daily records
    daily_rows = []
    for i, day_date in enumerate(dates):
        daily_row = row.copy()
        daily_row['day'] = day_date.day
        daily_row['full_date'] = day_date
        daily_row['day_of_week'] = day_date.dayofweek
        daily_row['is_weekend'] = day_date.dayofweek >= 5
        daily_row['significant_delay'] = i in day_indices

        if i in day_indices:
            position = sorted(day_indices).index(i)
            daily_row['delay_proxy'] = int(10 * (position + 1) / len(day_indices))
        else:
            daily_row['delay_proxy'] = 1

        daily_rows.append(daily_row)

    return daily_rows

# Generating daily delay records
print("Generating daily delay data...")
daily_rows = []
for _, row in tqdm(delay_df.iterrows(), total=len(delay_df)):
    daily_rows.extend(distribute_to_daily(row))

daily_delays_df = pd.DataFrame(daily_rows)
print(f"Generated {len(daily_delays_df)} daily delay records from {len(delay_df)} monthly records")

# Displaying class balance
class_balance = daily_delays_df['significant_delay'].value_counts(normalize=True) * 100
print("\nClass balance:")
print(f"Non-significant delays: {class_balance[False]:.1f}%")
print(f"Significant delays: {class_balance[True]:.1f}%")

# Displaying sample rows
print("\nSample of daily delays with target variable:")
print(daily_delays_df[['line', 'full_date', 'day_of_week', 'is_weekend', 'delay_proxy', 'significant_delay']].head(10))

# Connecting delays to stations using line mappings
print("\nConnecting delay data to stations...")

# Creating mapping between stations and lines
station_lines_df = []
for station, lines in station_to_lines_mapping.items():
    for line in lines:
        station_lines_df.append({
            'station_complex_standardized': station,
            'line': line
        })
station_lines_df = pd.DataFrame(station_lines_df)

# Merging daily delays with station lines
station_daily_delays = station_lines_df.merge(
    daily_delays_df,
    on='line',
    how='left'
)

# Aggregating delays per station per day
station_daily_delays_agg = station_daily_delays.groupby(
    ['station_complex_standardized', 'full_date']
).agg({
    'delay_proxy': ['sum', 'mean', 'max'],
    'significant_delay': 'any'
}).reset_index()

# Flattening column names
station_daily_delays_agg.columns = ['_'.join(col).strip('_') if isinstance(col, tuple) else col for col in station_daily_delays_agg.columns]

# Renaming columns for clarity
station_daily_delays_agg = station_daily_delays_agg.rename(columns={
    'delay_proxy_sum': 'delays_sum',
    'delay_proxy_mean': 'delays_mean',
    'delay_proxy_max': 'delays_max',
    'significant_delay_any': 'significant_delay_any'
})

# Adding weekday and weekend indicators
station_daily_delays_agg['day_of_week'] = pd.to_datetime(station_daily_delays_agg['full_date']).dt.dayofweek
station_daily_delays_agg['is_weekend'] = station_daily_delays_agg['day_of_week'] >= 5

# Merging with demographics
station_daily_delays_with_demographics = station_daily_delays_agg.merge(
    station_metrics_with_demographics,
    on='station_complex_standardized',
    how='left'
)

# Displaying summary
print(f"Created station-level daily delay dataset with {len(station_daily_delays_agg)} records")
print(f"After merging with demographics: {len(station_daily_delays_with_demographics)} records")
print("Sample of station daily delays with demographics:")
print(station_daily_delays_with_demographics.head())

# Checking for missing values
print("\nChecking for missing values after merging:")
for col in ['delays_sum', 'significant_delay_any', 'median_income', 'pct_public_transit']:
    missing = station_daily_delays_with_demographics[col].isna().sum()
    pct_missing = (missing / len(station_daily_delays_with_demographics)) * 100
    print(f"  {col}: {missing} missing ({pct_missing:.1f}%)")

# Creating income quartiles
station_metrics_with_demographics['income_quartile'] = pd.qcut(
    station_metrics_with_demographics['median_income'],
    4,
    labels=['Low', 'Medium-Low', 'Medium-High', 'High']
)

# Creating majority minority flag
station_metrics_with_demographics['majority_minority'] = station_metrics_with_demographics['pct_white'] < 50

# Saving datasets
print("\nSaving final datasets for future modeling and analysis...")

output_folder = "datasets/processed"

daily_delays_df.to_csv(f"{output_folder}/line_daily_delays.csv", index=False)
station_daily_delays_agg.to_csv(f"{output_folder}/station_daily_delays.csv", index=False)
station_daily_delays_with_demographics.to_csv(f"{output_folder}/station_daily_delays_with_demographics.csv", index=False)
station_metrics_with_demographics.to_csv(f"{output_folder}/station_metrics_with_demographics.csv", index=False)

# Displaying summary of saved files
print(f"Generated {len(daily_delays_df)} daily delay records")
print(f"Created {len(station_daily_delays_with_demographics)} station-day records with demographic data")
print("\nDatasets saved to:", output_folder)
```

```python
# Step 9: Creating Specialized Datasets for Modeling and Analysis

print("Step 9: Creating specialized datasets for modeling and analysis")
print("=" * 80)

# Creating training dataset
print("Creating training dataset...")

# Copying base daily delays
training_df = daily_delays_df.copy()

# Extracting temporal features
training_df['year'] = pd.to_datetime(training_df['full_date']).dt.year
training_df['month'] = pd.to_datetime(training_df['full_date']).dt.month
training_df['day_of_month'] = pd.to_datetime(training_df['full_date']).dt.day

# Cleaning division labels or inferring them
if 'division' in training_df.columns:
    training_df['division_clean'] = training_df['division'].str.replace(' DIVISION', '').str.strip()
elif 'division_mapped' in training_df.columns:
    training_df['division_clean'] = training_df['division_mapped']
else:
    def infer_division(line):
        try:
            int(line)
            return 'A'
        except:
            return 'B'
    training_df['division_clean'] = training_df['line'].apply(infer_division)

# Sorting records by line and date
training_df = training_df.sort_values(['line', 'full_date'])

# Creating lagged features using significant delay
training_df['significant_delay_int'] = training_df['significant_delay'].astype(int)
training_df['prev_day_delay'] = training_df.groupby('line')['significant_delay_int'].shift(1)
training_df['prev_week_delay'] = training_df.groupby('line')['significant_delay_int'].shift(7)
training_df['delay_avg_7day'] = training_df.groupby('line')['significant_delay_int'].transform(
    lambda x: x.rolling(7, min_periods=1).mean()
)

# Removing temporary conversion column
training_df.drop(columns=['significant_delay_int'], inplace=True)

# Creating day-of-week indicators
training_df['is_monday'] = (training_df['day_of_week'] == 0).astype(int)
training_df['is_friday'] = (training_df['day_of_week'] == 4).astype(int)
training_df['is_saturday'] = (training_df['day_of_week'] == 5).astype(int)
training_df['is_sunday'] = (training_df['day_of_week'] == 6).astype(int)

# Creating season feature
training_df['season'] = pd.cut(
    training_df['month'],
    bins=[0, 3, 6, 9, 12],
    labels=['Winter', 'Spring', 'Summer', 'Fall'],
    include_lowest=True
)

# Adding reporting category dummies if available
if 'reporting_category' in training_df.columns:
    reporting_dummies = pd.get_dummies(
        training_df['reporting_category'],
        prefix='cat',
        dummy_na=False
    )
    training_df = pd.concat([training_df, reporting_dummies], axis=1)

# Selecting final columns
model_features = [
    'line', 'full_date', 'division_clean',
    'significant_delay', 'delay_proxy',
    'year', 'month', 'day_of_month', 'day_of_week', 'is_weekend',
    'is_monday', 'is_friday', 'is_saturday', 'is_sunday', 'season',
    'reporting_category', 'subcategory', 'day_type',
    'prev_day_delay', 'prev_week_delay', 'delay_avg_7day',
]

# Appending dummy columns if available
if 'reporting_category' in training_df.columns:
    for col in reporting_dummies.columns:
        model_features.append(col)

# Retaining only available columns
available_columns = [col for col in model_features if col in training_df.columns]
training_df = training_df[available_columns]

# Filling missing values
for col in training_df.columns:
    if pd.api.types.is_categorical_dtype(training_df[col]):
        if training_df[col].isna().any():
            most_common = training_df[col].mode()[0]
            training_df[col] = training_df[col].fillna(most_common)
    elif col in ['prev_day_delay', 'prev_week_delay', 'delay_avg_7day']:
        training_df[col] = training_df[col].fillna(0)
    elif pd.api.types.is_numeric_dtype(training_df[col]):
        training_df[col] = training_df[col].fillna(training_df[col].median())
    else:
        if training_df[col].isna().any():
            most_common = training_df[col].mode()[0]
            training_df[col] = training_df[col].fillna(most_common)

# Creating analysis dataset
print("\nCreating analysis dataset...")
analysis_df = station_daily_delays_with_demographics.copy()

# Creating station-level delay statistics
station_delay_stats = analysis_df.groupby('station_complex_standardized').agg({
    'delays_sum': ['mean', 'median', 'max', 'std', 'count'],
    'significant_delay_any': 'mean',
    'is_weekend': ['mean', 'count']
}).reset_index()

# Flattening column names
station_delay_stats.columns = ['_'.join(col).strip('_') if isinstance(col, tuple) else col for col in station_delay_stats.columns]

# Renaming for clarity
station_delay_stats = station_delay_stats.rename(columns={
    'delays_sum_mean': 'avg_daily_delays',
    'delays_sum_median': 'median_daily_delays',
    'delays_sum_max': 'max_daily_delays',
    'delays_sum_std': 'std_daily_delays',
    'delays_sum_count': 'days_with_data',
    'significant_delay_any_mean': 'pct_days_with_significant_delay',
    'is_weekend_mean': 'pct_weekend_data',
    'is_weekend_count': 'total_days'
})

# Merging delay stats with demographics
analysis_df = station_metrics_with_demographics.merge(
    station_delay_stats,
    on='station_complex_standardized',
    how='left'
)

# Creating income categories
if 'income_quartile' in analysis_df.columns:
    analysis_df['income_category'] = analysis_df['income_quartile']
else:
    analysis_df['income_category'] = pd.qcut(
        analysis_df['median_income'],
        4,
        labels=['Low', 'Medium-Low', 'Medium-High', 'High']
    )

# Creating racial composition categories
def get_dominant_race(row):
    race_cols = ['pct_white', 'pct_black', 'pct_asian', 'pct_hispanic']
    race_labels = ['Predominantly White', 'Predominantly Black', 'Predominantly Asian', 'Predominantly Hispanic']
    for i, col in enumerate(race_cols):
        if row[col] > 50:
            return race_labels[i]
    return "Diverse"

analysis_df['racial_composition'] = analysis_df.apply(get_dominant_race, axis=1)

# Creating car dependency categories
analysis_df['car_dependency'] = pd.cut(
    analysis_df['pct_no_vehicle'],
    bins=[0, 25, 50, 75, 100],
    labels=['High Car Dependency', 'Moderate Car Dependency', 'Low Car Dependency', 'Car Independent'],
    include_lowest=True
)

# Creating transit usage categories
analysis_df['transit_usage'] = pd.cut(
    analysis_df['pct_public_transit'],
    bins=[0, 20, 40, 60, 100],
    labels=['Low Transit Usage', 'Moderate Transit Usage', 'High Transit Usage', 'Very High Transit Usage'],
    include_lowest=True
)

# Calculating equity metrics
analysis_df['delay_per_ridership'] = analysis_df['avg_daily_delays'] / analysis_df['ridership_mean'].replace(0, 1)
analysis_df['delay_impact_score'] = analysis_df['avg_daily_delays'] * analysis_df['pct_days_with_significant_delay']
analysis_df['transit_dependency_ratio'] = analysis_df['pct_no_vehicle'] * analysis_df['pct_public_transit'] / 100
analysis_df['delay_equity_ratio'] = analysis_df['delay_impact_score'] / (analysis_df['median_income'] / 50000)

# Calculating weekend vs weekday patterns
if 'is_weekend' in analysis_df.columns:
    weekend_weekday = station_daily_delays_with_demographics.groupby(
        ['station_complex_standardized', 'is_weekend']
    )['delays_sum'].mean().unstack()
    if 0 in weekend_weekday.columns and 1 in weekend_weekday.columns:
        weekend_weekday.columns = ['weekday_delays', 'weekend_delays']
        weekend_weekday['weekend_weekday_ratio'] = weekend_weekday['weekend_delays'] / weekend_weekday['weekday_delays'].replace(0, 1)
        analysis_df = analysis_df.merge(
            weekend_weekday.reset_index(),
            on='station_complex_standardized',
            how='left'
        )

# Handling missing or infinite values
print("\nChecking for missing or infinite values...")
for col in analysis_df.select_dtypes(include=['float64']).columns:
    inf_mask = np.isinf(analysis_df[col])
    if inf_mask.any():
        inf_count = inf_mask.sum()
        print(f"Found {inf_count} infinity values in {col}")
        valid_values = analysis_df[col][~np.isinf(analysis_df[col])]
        if len(valid_values) > 0:
            median_val = valid_values.median()
            analysis_df[col].replace([np.inf, -np.inf], median_val, inplace=True)
            print(f"  Replaced with median value: {median_val}")
    na_mask = analysis_df[col].isna()
    if na_mask.any():
        na_count = na_mask.sum()
        print(f"Found {na_count} NaN values in {col}")
        analysis_df[col].fillna(analysis_df[col].median(), inplace=True)

# Displaying summary
print(f"\nCreated training dataset with {len(training_df)} records and {len(training_df.columns)} variables")
print(f"Created analysis dataset with {len(analysis_df)} records and {len(analysis_df.columns)} variables")

# Saving datasets
print("\nSaving specialized datasets...")
training_df.to_csv(f"{output_folder}/training_dataset.csv", index=False)
analysis_df.to_csv(f"{output_folder}/analysis_dataset.csv", index=False)

print(f"Saved training dataset to {output_folder}/training_dataset.csv")
print(f"Saved analysis dataset to {output_folder}/analysis_dataset.csv")
```
